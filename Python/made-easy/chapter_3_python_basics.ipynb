 ####  3.4.4. Выполнение выражений  

 **Выражение** - совокупность значений, переменных и операторов. (Не обязательно все сразу). Вычислить выражение - это не то же самое, что вывести его значение на экран.  

 >   множественное присваивание: переменные а и b  одновременно получают значения о и 1:  a, b = 0, 1  
 > перед таким присваиванием всегда сначала вычисляется правая часть: a, b = b, a + b  
 > Цикл while выполняется до тех пор, пока условие остается истинным. В Python ис­тинным является любое ненулевое целочисленное значение, ноль считается ложным. Условие также может быть строкой или списком и вообще любой по­следовательностью. При этом все, что имеет ненулевую длину, считается истин­ным, а пустые последовательности - ложными. стандартные операторы сравнения: < (меньше), > (больше), == (равно), <= (меньше или равно), >= (больше или равно) и ! = (не равно)  
 > Тело цикла записано с отступом. **Отступы** - это способ группировки выраже­ний в Python. В интерактивном режиме интерпретатора для задания отступа нужно использовать табуляцию или пробелы. На практике код бывает гораздо более сложным, и у всех специализированных текстовых редакторов есть функ­ция автоматического отступа. Когда составное выражение вводится в интерак­тивном режиме, за ним должна следовать пустая строка, указывающая на завер­шение (поскольку синтаксический анализатор не может угадать, когда блок кода закончился). У всех строк в блоке должен быть одина­ковый отступ.  
 > Функция **print()** выводит значение переданного аргумента на экран. Строки выводятся без кавычек, а между аргументами вставляется пробел, что позволяет красиво форматировать результаты

   #### 3.5.1.  Функции print()

 - **Аргументы(args)** - все то, что мы передаем в функцию. Например, аргументы - строка, класс или переменные, которые мы передаем в функцию print()  
 ***args** - передать функции любое количество переменных (можно *objects)  
 - **именованные аргументы(rwargs)**  - аргументы, для которых ппри передаче в функцию указывается не только значение, но и имя.  

 sep='' sep - именованный аргумент, у него есть имя, справа от символа = указано присвоенное ему значение.  

 **Значение по умолчанию** - значение, указанное в определении функции. Если не занзачить аргументу другое значение, будет передано значение по умолчанию, синтаксических ошибок не возникнет.  

 **Параметры функции print()**  

 - *objects  - любое количество объектов. Перед выводом все объекты собираются в строку.  

 - sep='' (необязательный) - задает разделитель объекто, если их несколько. Значение по умолчанию - символ пробела, указанный как ' '  
 - end=' ' (необязательный) - определяет, что вывести в конце строки. Значение по умолчанию - символ переноса строки '\n'  
 - filr=sys.stdout (необязательный) - объект с методом write(string). Если этот парамет не указан, будет использоваться sys.stdout по умолчанию, что означает вывод результатов на экран.  
 - flush=False (необязательный) - логическое заявление, указывающее, будет ли вывод очищен (True) или буферезирован (False). По умолчанию - False

###  3.6. **Основные виды ошибок**  
- Синтаксическая ошибка - проблемы с языковыми конструкциями  
- Ошибка времени выполнения - проблемы с выполнением кода.  
- Семантическая ошибка - неожиданный результат.  

#### 3.6.1. Синтаксические ошибки  

Вызваны опечатками или возникла проблема со структурой программы. Не соблюдается синтаксис Python. 
`In[35]: # отсутствует закрывающая кавычка
        print("Hello World!)  
File "<ipython-input-49-9d0e3bd45f27>", line 3  
        print ("Hello World! )  

SyntaxError: EOL while scanning string literal  

Указано место ошибки  

#### 3.6.2.  Ошибки времени выполнения  

Возникают во время работы программы. Поскольку Python является интерпретируемым языком, такие ошибки возникают лишь в тот момент, когда выполнение программы доходит до ошибочной строки.  

**Причины ошибок времени выполнения**  

- неверно введенное имя переменной или функции  
- использование переменной до ее определения  
- имя должно быть заключено в кавычки  
- деление на ноль  

Ошибка времени выполнения возникает, когда Python понимает саму команду, но при ее выполнении сталкивается с проблемами. Поэтому эта ошибка и называется «времени выполнения», поскольку возникает только после запуска программы.  

`In[36]: 5/0`  

ZeroDivisionError Traceback (most recent call last) <ipython-input-51-0106664d39e8> in <module>
----> 1 5/0
ZeroDivisionError: division Ьу zero  

####  3.6.3. Семантические ошибки  

**Семантические или логические ошибки**- это проблемы с самим построением вашей программы. Обычно такие проблемы не вызывают сообщений об ошибках, но поведение программы оказывается неверным.  

Эти ошибки часто бывают вызваны случайным использованием неверных пере­ менных либо просто неправильными вычислениями.  

`In[37]: # вывести "Hello Nilabh"`  
        `name = "Nilabh"`  
        `print("Hello name")`  
Hello name  

#### 3.6.4. Альтернативные методы устранения ошибок  

- Поиск в интернете stackoverflow.com  
- Избегать таких случаев, писать по несколько строк кода, а потом выполнять программу.  
- Промежуточные версии. Сохранять промежуточный промежуточный прогресс в разра­ботке программы каждый раз, когда начинаем  писать новую часть. Если достигли определенного этапа, сделайть резервную копию. Для этого достаточ­ но выбрать пункт меню **Save as** и назвать свой файл **filename2.py**, затем **filename3.py** и т. д. Если вы работаете с filename7.py и столкнулись с нерешае­мой проблемой, можно просто вернуться к filename6.py. По окончании работы можно просто убрать все старые файлы в архив и переименовать окончательный файл по своему усмотрению.  
- Комментарии. Закомментировать проблемный участок кода, чтобы посмотреть, сохранится ли проблема. Можно использовать воз­можности среды IDE, чтобы поместить символ комментария перед всеми стро­ками выделенного фрагмента кода. Изолировать разные участки кода, чтобы сузить область поиска проблемы.  
- Все сначала.  Начать писать код заново с нового пустого файла, затем скопировать и всавить небольшие фрагменты кода из старого файла по одному. Каждый раз, добавляя новый раздел, запускать программу и проверять, не появилась ли проблема.
